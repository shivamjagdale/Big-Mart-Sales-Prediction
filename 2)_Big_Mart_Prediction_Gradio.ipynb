# **Label Encoder 1**

```python
encoder1 = LabelEncoder()
```

```python
Item_Identifier = encoder1.fit_transform(df["Item_Identifier"])
```

```python
df["Item_Identifier"] = Item_Identifier
```

```python
df["Item_Weight"] = df["Item_Weight"].fillna(df["Item_Weight"].mean())
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('LF', 'Low Fat')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('low fat', 'Low Fat')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('reg', 'Regular')
```

```python
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("Low Fat", 0)
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("Regular", 1)
```

# **Label Encoder 2**

```python
encoder2 = LabelEncoder()
```

```python
Item_Type = encoder2.fit_transform(df["Item_Type"])
```

```python
df["Item_Type"] = Item_Type
```

# **Label Encoder 3**

```python
encoder3 = LabelEncoder()
```

```python
Outlet_Identifier	= encoder3.fit_transform(df["Outlet_Identifier"])
```

```python
df["Outlet_Identifier"] = Outlet_Identifier
```

```python
df["Outlet_Size"].value_counts()
```

```python
df["Outlet_Size"] = df["Outlet_Size"].replace("Small", 0)
df["Outlet_Size"] = df["Outlet_Size"].replace("Medium", 1)
df["Outlet_Size"] = df["Outlet_Size"].replace("High",2)
```

```python
df["Outlet_Location_Type"].value_counts()
```

```python
df["Outlet_Location_Type"] = df["Outlet_Location_Type"].replace("Tier 1", 0)
df["Outlet_Location_Type"] = df["Outlet_Location_Type"].replace("Tier 2", 1)
df["Outlet_Location_Type"] = df["Outlet_Location_Type"].replace("Tier 3", 2)
```

```python
df["Outlet_Type"].value_counts()
```

```python
df["Outlet_Type"] = df["Outlet_Type"].replace("Supermarket Type1", 0)
df["Outlet_Type"] = df["Outlet_Type"].replace("Grocery Store", 1)
df["Outlet_Type"] = df["Outlet_Type"].replace("Supermarket Type3", 2)
df["Outlet_Type"] = df["Outlet_Type"].replace("Supermarket Type2", 3)
```

```python
X = df.drop(columns='Item_Outlet_Sales')
```

```python
Y = df['Item_Outlet_Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn import metrics
```

```python
X_train_prediction = model.predict(X_train)
r_squared = metrics.r2_score(Y_train, X_train_prediction)

print(r_squared)
```

```python
X_test_prediction = model.predict(X_test)
r_squared = metrics.r2_score(Y_test, X_test_prediction)

print(r_squared)
```

```python
import gradio as gr
import numpy as np
import joblib

# Load model, scaler, and encoders
model = joblib.load("mart_model.pkl")
scaler = joblib.load("scaler_mart.pkl")
encoder1 = joblib.load("encoder1_mart.pkl")
encoder2 = joblib.load("encoder2_mart.pkl")
encoder3 = joblib.load("encoder3_mart.pkl")

# Get dropdown values from encoders
item_ids = encoder1.classes_.tolist()
item_types = encoder2.classes_.tolist()
outlet_ids = encoder3.classes_.tolist()

def predict_sales(item_id, weight, fat_content, visibility, item_type, mrp,
                  outlet_id, est_year, outlet_size, outlet_location, outlet_type):

    # Encode categorical inputs
    fat_encoded = 0 if fat_content == "Low Fat" else 1
    outlet_size_map = {"Small": 0, "Medium": 1, "High": 2}
    outlet_location_map = {"Tier 1": 0, "Tier 2": 1, "Tier 3": 2}
    outlet_type_map = {
        "Supermarket Type1": 0,
        "Grocery Store": 1,
        "Supermarket Type3": 2,
        "Supermarket Type2": 3
    }

    try:
        a = encoder1.transform([item_id])[0]
        b = encoder2.transform([item_type])[0]
        c = encoder3.transform([outlet_id])[0]
    except ValueError as e:
        return f"Error: {e}"

    # Final input array
    input_data = np.array([[a, weight, fat_encoded, visibility, b, mrp,
                            c, est_year, outlet_size_map[outlet_size],
                            outlet_location_map[outlet_location],
                            outlet_type_map[outlet_type]]])

    # Standardize and predict
    input_scaled = scaler.transform(input_data)
    prediction = model.predict(input_scaled)[0]

    return f"Predicted Sales: ₹{round(prediction, 2)}"

# Gradio UI
gr.Interface(
    fn=predict_sales,
    inputs=[
        gr.Dropdown(item_ids, label="Item Identifier"),
        gr.Number(label="Item Weight"),
        gr.Radio(["Low Fat", "Regular"], label="Item Fat Content"),
        gr.Number(label="Item Visibility"),
        gr.Dropdown(item_types, label="Item Type"),
        gr.Number(label="Item MRP"),
        gr.Dropdown(outlet_ids, label="Outlet Identifier"),
        gr.Number(label="Outlet Establishment Year"),
        gr.Radio(["Small", "Medium", "High"], label="Outlet Size"),
        gr.Radio(["Tier 1", "Tier 2", "Tier 3"], label="Outlet Location Type"),
        gr.Radio(["Supermarket Type1", "Grocery Store", "Supermarket Type3", "Supermarket Type2"], label="Outlet Type"),
    ],
    outputs=gr.Text(label="Predicted Item Outlet Sales"),
    title="🛒 Big Mart Sales Predictor",
    description="Fill in the item and outlet details to predict the sales amount."
).launch()
```
